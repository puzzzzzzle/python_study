```python
code = """class Animal:
    def run(self):
        print('running')

class Dog(Animal):
    def yelp(self):
        print('woof')
    def play(self):
        print('playing')

class Sleuth(Dog):
    def yelp(self):
        print('WOOF!')
    def hunt(self):
        print('hunting')

class SnifferDog(Dog):
    def search(self):
        print('searching')

class PoliceDog(Sleuth, SnifferDog):
    def patrol(self):
        print('patroling')

print(PoliceDog.__base__)
print(PoliceDog.__bases__)
print(PoliceDog.__mro__)
"""
code = compile(code, filename="<string>", mode="exec")
exec (code, globals(),{})
print(dis.dis(code))
```
